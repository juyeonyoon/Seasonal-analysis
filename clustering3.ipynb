```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#전체 데이터 사용해보기 1908 - 2021

path = "/content/drive/MyDrive/데이터분석기초/과제/데이터/"
temp = pd.read_csv(path + "기온_total.csv", header=0)
prec = pd.read_csv(path + "강수량_total.csv", header=0)
hum = pd.read_csv(path + "습도_total.csv", header=0)
```

```python
#총 114년 * 12개월 = 1368(관측값) 이어야 되는데 1354개 빠진 값이 있다. 
```

```python
temp = temp[["일시", "평균기온(℃)"]]
prec = prec[["일시", "강수량(mm)"]]
hum = hum[["일시", "평균습도(%rh)"]]
```

```python
temp_prec = pd.merge(temp, prec, on="일시")
wheather = pd.merge(temp_prec, hum, on="일시")
```

```python
#결측치 확인 
#기온 (1950년 9월 - 1953 11월) 25개
print(wheather["평균기온(℃)"].isnull().sum())
wheather[wheather["평균기온(℃)"].isnull()] 
```

```python
#습도 (1950년 9월 - 1953 11월) 25개
print(wheather["평균습도(%rh)"].isnull().sum())
wheather[wheather["평균습도(%rh)"].isnull()] 
```

```python
#결측값 대체하자 10% 미만일때 삭제 또는 대체 대체법은 어느 것이든 사용 가능 
```

```python
wheather["month"] = [a.split(".")[0] for a in wheather["일시"]]
wheather["year"] = [a.split(".")[1] for a in wheather["일시"]]
```

```python
wheather.head()
```

```python
#결측값 월 평균으로 계산 
#강수량
replace_prec = round(wheather[wheather["month"] == "May"]["강수량(mm)"].mean(),1)
wheather["강수량(mm)"].fillna(replace_prec, inplace=True)
wheather.iloc[1695]
```

```python
#기온
replace_tem = round(wheather[wheather["month"] == "Jan"]["평균기온(℃)"].mean(),1)
index = wheather[(wheather["month"]=="Jan")&(wheather["평균기온(℃)"].isnull())].index
wheather.loc[index, "평균기온(℃)"] = replace_tem
```

```python
for i in ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]:
    replace_tem = round(wheather[wheather["month"]==i]["평균기온(℃)"].mean(),1)
    replace_hum = round(wheather[wheather["month"]==i]["평균습도(%rh)"].mean(),1)
    index_tem = wheather[(wheather["month"]==i)&(wheather["평균기온(℃)"].isnull())].index
    index_hum = wheather[(wheather["month"]==i)&(wheather["평균습도(%rh)"].isnull())].index
    wheather.loc[index_tem, "평균기온(℃)"] = replace_tem
    wheather.loc[index_hum, "평균습도(%rh)" ] = replace_hum
```
